```python
import numpy as np
import matplotlib.pyplot as plt

A = [[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]]

A = np.array(A)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
#Уменьшить объект в два раза
b=np.array([[0.5,0,0],[0,0.5,0],[0,0,0.5]])
res=b@A.transpose()
tt=res.transpose()
tt
A = np.array(tt)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
#Повернуть на 130 градусов
#COS130 = -0.6428 по Байесу
#SIN13=0.766 по Байесу
#
s=np.array([[-0.6248,0.766,0],[-0.766,-0.6248,0],[0,0,1]])
_res=s@A.transpose()
_tt=_res.transpose()
_tt
A = np.array(_tt)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
#Отразить объект относительно прямой y=x
c=np.array([[0,1,0],[1,0,0],[0,0,1]])
cres=c@A.transpose()
ctt=cres.transpose()
A = np.array(ctt)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
total=np.linalg.det(b@s@c)
total
```
