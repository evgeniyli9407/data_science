```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
#1 Вычислить значения первой и второй производной функции в точке х=1:
x=sp.Symbol('x')
func=x**5+4*sp.sin(2*x)+sp.cos(3*x+3)
st=sp.diff(func, x).evalf(subs={'x':1}) #в первой
sd=sp.diff(st, x).evalf(subs={'x':1})# во второй
```

```python
#2. Постройте график функции, первой и второй производной на отрезке[-5:5]
func_2 = sp.sin(2*x+1)**5
a=[]
b=[]
for i in range(-5,5):
    fst = sp.diff(func_2, x).evalf(subs={'x':i})
    snd = sp.diff(fst, x).evalf(subs={'x':i})
    a.append(fst)
    b.append(snd)

plt.plot(a)
plt.plot(b)
```

```python
#задание №3 Вообще не понял задание и для чего оно нужно, перечитал, пытался вникнуть-тщетно
import numpy as np
import matplotlib.pyplot as plt
def relu(x):
    return np.maximum(0.0, x)
x = np.arange(-3, 3, 0.1)
y = relu(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('ReLU Function')
plt.show()
```

```python
def relu_derivative(x):
    # Вычислить производную функции ReLU
    # Ваш код напишите ниже и результат передайте в переменную derivative
    derivative = sp.diff(relu(x))
    return derivative
```

```python
#Задание 4 Вообще не понял задание и для чего оно нужно, перечитал, пытался вникнуть-тщетно
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
x = np.arange(-5, 5, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sigmoid Function')
plt.show()
def sigmoid_derivative(x):
    # Вычислить производную сигмоидной функции
    # Ваш код напишите ниже и результат передайте в переменную derivative
    derivative = sp.diff(sigmoid(x))
    return derivative
```
